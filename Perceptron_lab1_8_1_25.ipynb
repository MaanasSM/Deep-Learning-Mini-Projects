```python
# AND GATE

import numpy as np                 #for numerical operation
import matplotlib.pyplot as plt    #for plotting


# Input data and targets
X = np.array([[1, 1], [1, -1], [-1, 1], [-1, -1]])  # Input data (bipolar inputs) #00,01,10,11 = -1-1,-11,1-1,11
y = np.array([1, -1, -1, -1])  # Target outputs (bipolar targets)


# Perceptron initialization
weights = np.zeros(3)  # 2 inputs + 1 bias  # intialization
learning_rate = 0.1    # in image, for learning we chumma took learning rate as 1, for programming we keep as 0.1
epochs = 2  # Set to 2 epochs to force convergence at epoch 2


# Training loop
for epoch in range(epochs):
    total_error = 0  # Track total error during the epoch  # error = 0 => output = target (when total error  = 0, we decide if any more epochs are to be done or not)
    for i in range(len(X)):
        x = np.append(X[i], 1)  # Add bias input (1)
        prediction = np.sign(np.dot(x, weights))  # Prediction using the sign activation function
        error = y[i] - prediction  # Compute error
        weights += learning_rate * error * x  # Update weights
        total_error += abs(error)  # Accumulate error


    # Check for convergence: If no error, break out early
    if total_error == 0:
        print(f"Converged at epoch {epoch + 1}")
        break
    else:
        print(f"Epoch {epoch + 1} - Total Error: {total_error}")


# After training, calculate accuracy
predictions = np.sign(np.dot(np.c_[X, np.ones(len(X))], weights))  # Make predictions
accuracy = np.mean(predictions == y) * 100  # Calculate accuracy
print(f"Accuracy after {epoch + 1} epoch(s): {accuracy:.2f}%")


# Plotting decision boundary
x1_vals = np.linspace(-2, 2, 100)
x2_vals = -(weights[0] * x1_vals + weights[2]) / weights[1]  # Calculate decision boundary


# Visualize data points and decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', label="Data points")
plt.plot(x1_vals, x2_vals, color='black', label="Decision boundary")
plt.legend()
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron for AND function")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Training data and target values
X = np.array([[1, 1, 1, 1],   # x1
              [-1, 1, -1, -1], # x2
              [1, 1, 1, -1],   # x3
              [1, -1, -1, 1]]) # x4
y = np.array([1, 1, -1, -1])  # Target values


# Initialize weights (including the bias)
weights = np.zeros(X.shape[1])  # 4 inputs (x1, x2, x3, x4)
learning_rate = 1


# Training the perceptron
for epoch in range(10):  # We can run for a few epochs
    print(f"Epoch {epoch + 1}")
    for i in range(len(X)):
        x = X[i]
        target = y[i]

        # Calculate the perceptron output
        net_input = np.dot(x, weights)
        prediction = 1 if net_input >= 0 else -1

        # If prediction is incorrect, update weights
        if prediction != target:
            weights += learning_rate * (target - prediction) * x
            print(f"Updated weights: {weights}")
        else:
            print(f"Correct prediction: {prediction} for input {x}")

    print(f"End of epoch {epoch + 1}, current weights: {weights}")


# Calculate accuracy
correct_predictions = 0
for i in range(len(X)):
    x = X[i]
    net_input = np.dot(x, weights)
    prediction = 1 if net_input >= 0 else -1
    if prediction == y[i]:
        correct_predictions += 1


accuracy = (correct_predictions / len(X)) * 100
print(f"\nAccuracy: {accuracy:.2f}%")


# Testing the trained perceptron on the input vectors
print("\nFinal weights:", weights)
print("\nTesting the perceptron response on the training vectors:")
for i in range(len(X)):
    x = X[i]
    net_input = np.dot(x, weights)
    prediction = 1 if net_input >= 0 else -1
    print(f"Input {x} -> Predicted: {prediction}, Target: {y[i]}")


# Plotting decision boundary
# Since we have 4 inputs, we can't plot in 2D directly. Let's visualize the 2D projection.
# Assuming the first two inputs (x1 and x2) determine the decision boundary.


x1_vals = np.linspace(-2, 2, 100)
x2_vals = np.linspace(-2, 2, 100)


# Create a meshgrid for plotting
X1, X2 = np.meshgrid(x1_vals, x2_vals)


# Calculate corresponding x3 and x4 from the decision boundary equation
# w1 * x1 + w2 * x2 + w3 * x3 + w4 * x4 = 0
# Here we need to assume a relation, for example, solving for x4 based on x1, x2, and the weights.


# Solve for x4 given x1 and x2 (assuming x3 is fixed as 1)
x3 = 1
x4_vals = (-weights[0] * X1 - weights[1] * X2 - weights[2] * x3) / weights[3]


# Plot the decision boundary
plt.figure(figsize=(8, 6))
plt.contour(X1, X2, x4_vals, levels=[0], colors='black', label="Decision Boundary")


# Plotting data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', label="Data points")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron for Binary Classification (Class 1 and Class -1)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Input data and target values for the AND function (bipolar inputs and targets)
X = np.array([[1, 1],   # x1
              [1, -1],  # x2
              [-1, 1],  # x3
              [-1, -1]]) # x4
y = np.array([1, -1, -1, -1])  # Target values for AND function


# Initialize weights to zero (including bias)
weights = np.zeros(2)  # No bias term in this example; only two inputs
learning_rate = 0.1     # Learning rate (adjustable)
epochs = 10             # Number of epochs to run


# Hebbian Learning Rule
for epoch in range(epochs):
    for i in range(len(X)):
        x = X[i]
        target = y[i]

        # Update weights based on Hebbian rule
        weights += learning_rate * target * x

    print(f"Epoch {epoch + 1}: Weights: {weights}")


# After training, calculate predictions
predictions = np.sign(np.dot(X, weights))  # Apply the sign function to get predicted class
accuracy = np.mean(predictions == y) * 100  # Calculate accuracy


# Output results
print(f"\nAccuracy after {epochs} epochs: {accuracy:.2f}%")


# Plotting decision boundary
# The decision boundary is the line where the perceptron output changes (i.e., when the weighted sum is 0)
x1_vals = np.linspace(-2, 2, 100)
x2_vals = -(weights[0] * x1_vals) / weights[1]  # Equation of decision boundary


# Plotting data points and decision boundary
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', label="Data points")
plt.plot(x1_vals, x2_vals, color='black', label="Decision boundary")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Hebbian Learning for AND Function")
plt.legend()
plt.grid(True)
plt.show()
```
